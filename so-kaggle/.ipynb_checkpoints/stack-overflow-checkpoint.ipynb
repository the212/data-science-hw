```python
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import metrics
```

```python
questions = pd.read_csv('train.csv')
```

```python
def cleanDF(df):
    df['OpenStatus'] = df['OpenStatus'].apply(lambda x: 0 if x == 'open' else 1)
    titles = df[['Id', 'Title', 'OpenStatus']]
    
    return titles
```

```python
train_data = cleanDF(questions)
```

```python
X = train_data.Title
y = train_data.OpenStatus
```

```python
count_vect = CountVectorizer(stop_words='english')
# yeields ~0.2 log loss count_vect = CountVectorizer(stop_words='english', ngram_range=(1, 2))
# yeields ~0.3 log loss count_vect = CountVectorizer(stop_words='english', ngram_range=(1, 3))

count_vect.fit(X)
X_counts = count_vect.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_counts, y.values)
```

```python
clf = MultinomialNB()
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
y_pred_prod = clf.predict_proba(X_test)
```

```python
metrics.log_loss(y_test, y_pred_prod)
```

```python
test_data = pd.read_csv('test.csv')
```

```python
X = test_data.Title
```

```python
X_counts = count_vect.transform(X)
```

```python
test_predictions = clf.predict_proba(X_counts)
```

```python
test_data['Prediction'] = test_predictions[:,1]
```
